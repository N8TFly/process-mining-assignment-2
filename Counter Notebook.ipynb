```python
def columnMaxCounter(data, groupByColumn):
    """Returns a DF with per case ID - each Activity and its max counter at a certain time"""
    #FIXME: activities = DF
    if groupByColumn == 'Sub_Process':
        counterCol = 'cum_subprocess' 
    elif groupByColumn == 'activity':
        counterCol = 'cum_activity' 
    
    time_per_doctype = data.groupby(['Case ID', groupByColumn], as_index=True).agg(
        counter = (counterCol, 'max'))
    time_per_doctype.reset_index(inplace = True)
    print(time_per_doctype)
    caseOccurences = time_per_doctype.pivot(index='Case ID', columns=groupByColumn, values=['counter']).reset_index()
    caseOccurencesColumnNames = [col[1] + "_" + col[0] if col[0] != 'Case ID' else col[0] for col in caseOccurences.columns.to_list()]

    caseOccurences.columns = caseOccurencesColumnNames
    caseOccurencesColumnNames.sort()
    caseOccurences = caseOccurences[caseOccurencesColumnNames] #rearrange

    if groupByColumn == 'Sub_Process':
        missingCols = list(set(counter_doctype_sub_all_collumns) - set(caseOccurencesColumnNames))
    elif groupByColumn == 'activity':
        missingCols = list(set(counter_activities_all_columns) - set(caseOccurencesColumnNames))

    #caseOccurences = caseOccurences.reindex(columns=caseOccurences.columns.tolist() + missingCols)
    caseOccurences = caseOccurences.reindex(columns=caseOccurences.columns.tolist())
    return caseOccurences
```

### Make the output

```python
def find_date_limit(cutoff, df):
    df = df[df['StartDate'].dt.year == 2016]
    df = df[df['eventcounter'] <= cutoff]
    df = df[df['eventcounter'] == df.eventcounter.max()]
    smallestdate = df['Complete Timestamp'].min()
    return smallestdate
    
```

```python
def setActivityCutoffPointTrain(cutoff):
    df1 = deepcopy(df)

#     filter out events later than cutoff point
#     CASE ID IN df[df['Event Log Counter'] == cutoff]['Case ID']
    df1 = df1[df1['eventcounter'] <= cutoff]
    datelimit = find_date_limit(cutoff, df1)
    
    dfResults = pd.DataFrame(df1[(df1['eventcounter'] <= cutoff) & (df1['Complete Timestamp'] < datelimit)].groupby('Case ID', as_index=True).agg(
        StartDate=('StartDate', 'first'),
        CaseEndDate=('EndDate', 'first'),
        Duration=('Total_Duration', 'last'),
        Department=('(case) department', 'first'),
        NumberParcels=('(case) number_parcels', 'first'),
        Area=('(case) area', 'first'),
        Year=('(case) year', 'first'),
        UO2= ('Undesired Outcome 2', 'first'),
        UO1= ('Undesired Outcome 1', 'first'),
        Normal = ('Normal', 'first'),
        AmountApplied = ('(case) amount_applied0', 'first'),
        Redistribution= ('(case) redistribution', 'first'),
        SmallFarmer = ('(case) small farmer', 'first'),
        YoungFarmer = ('(case) young farmer', 'first'),
        Current_SuccessRate=('succes_rate', 'last'),
        Current_TimeFromStart = ('TimePassed', 'last'),
        Max_lag = ('Max_lag', 'last'),
        Avg_lag = ('Avg_lag', 'last'),
        Applied_per_parcel = ('avg_amount_applied0_per_parcel', 'first'),
        Applied_per_area = ('avg_amount_applied0_per_area', 'first'),
        Current_Date = ('Complete Timestamp', 'last'),
        Area_per_parcel = ('avg_area_per_parcel', 'first'),
        StartMonth = ('StartMonth', 'first'),
        last_event_nr = ('eventcounter', 'max')))
    dfResults.reset_index(inplace = True)
    #TODO: Voor Undesired Outcome1 add UO1 = ('undesiredOutcome1', 'first'),
    #FIXME: Wij doen uiteindelijk niets met de andere amount1,2,3 etc. omdat we geen duidelijk punt hebben ontdekt dat die beschrijft.

    #One hot encode Department
    dfResults = pd.concat([dfResults, pd.get_dummies(dfResults['Department'], prefix='Department')], axis=1)
    dfResults.drop(['Department'], axis=1, inplace=True)
    dfResults = pd.concat([dfResults, pd.get_dummies(dfResults['StartMonth'], prefix='StartMonth')], axis=1)
    dfResults.drop(['StartMonth'], axis=1, inplace=True)
    

    #Get Max value counts
    doctype_subproccess_value_counts = columnMaxCounter(df1, groupByColumn='Sub_Process')
    doctype_subproccess_value_counts = doctype_subproccess_value_counts.add_prefix("DS ")
    activity_value_counts = columnMaxCounter(df1, groupByColumn='activity')
    activity_value_counts = activity_value_counts.add_prefix("Act ")
    #Merge the Dataframes
    dfResults = dfResults.merge(doctype_subproccess_value_counts, left_on='Case ID', right_on='DS Case ID', suffixes=(False, False))
    dfResults = dfResults.merge(activity_value_counts, left_on='Case ID', right_on='Act Case ID', suffixes=(False, False))

    dfResults = dfResults.fillna(0)


#     filterout = len(dfResults[dfResults['last_event_nr']!=cutoff])
#     print("Filterout {} cases because they have less than {} events".format(filterout, cutoff))
    dfResults = dfResults[dfResults['last_event_nr']<=cutoff]
    dfResults.drop(columns=['DS Case ID', 'Act Case ID', 'last_event_nr'], axis=1, inplace=True)
    dfResults = dfResults[dfResults['StartDate'].dt.year == 2015]
   

    return dfResults
```

```python
def setActivityCutoffPointTest(cutoff):
    df1 = deepcopy(df)

#     filter out events later than cutoff point
#     CASE ID IN df[df['Event Log Counter'] == cutoff]['Case ID']
    df1 = df1[df1['eventcounter'] <= cutoff]
        
    dfResults = pd.DataFrame(df1[df1['eventcounter'] <= cutoff].groupby('Case ID', as_index=True).agg(
        StartDate=('StartDate', 'first'),
        CaseEndDate=('EndDate', 'first'),
        Duration=('Total_Duration', 'last'),
        Department=('(case) department', 'first'),
        NumberParcels=('(case) number_parcels', 'first'),
        Area=('(case) area', 'first'),
        Year=('(case) year', 'first'),
        UO2= ('Undesired Outcome 2', 'first'),
        UO1= ('Undesired Outcome 1', 'first'),
        Normal = ('Normal', 'first'),
        AmountApplied = ('(case) amount_applied0', 'first'),
        Redistribution= ('(case) redistribution', 'first'),
        SmallFarmer = ('(case) small farmer', 'first'),
        YoungFarmer = ('(case) young farmer', 'first'),
        Current_SuccessRate=('succes_rate', 'last'),
        Current_TimeFromStart = ('TimePassed', 'last'),
        Max_lag = ('Max_lag', 'last'),
        Avg_lag = ('Avg_lag', 'last'),
        Applied_per_parcel = ('avg_amount_applied0_per_parcel', 'first'),
        Applied_per_area = ('avg_amount_applied0_per_area', 'first'),
        Current_Date = ('Complete Timestamp', 'last'),
        Area_per_parcel = ('avg_area_per_parcel', 'first'),
        StartMonth = ('StartMonth', 'first'),
        last_event_nr = ('eventcounter', 'max')))
    dfResults.reset_index(inplace = True)
    #TODO: Voor Undesired Outcome1 add UO1 = ('undesiredOutcome1', 'first'),
    #FIXME: Wij doen uiteindelijk niets met de andere amount1,2,3 etc. omdat we geen duidelijk punt hebben ontdekt dat die beschrijft.

    #One hot encode Department
    dfResults = pd.concat([dfResults, pd.get_dummies(dfResults['Department'], prefix='Department')], axis=1)
    dfResults.drop(['Department'], axis=1, inplace=True)
    dfResults = pd.concat([dfResults, pd.get_dummies(dfResults['StartMonth'], prefix='StartMonth')], axis=1)
    dfResults.drop(['StartMonth'], axis=1, inplace=True)
    

    #Get Max value counts
    doctype_subproccess_value_counts = columnMaxCounter(df1, groupByColumn='Sub_Process')
    doctype_subproccess_value_counts = doctype_subproccess_value_counts.add_prefix("DS ")
    activity_value_counts = columnMaxCounter(df1, groupByColumn='activity')
    activity_value_counts = activity_value_counts.add_prefix("Act ")
    #Merge the Dataframes
    dfResults = dfResults.merge(doctype_subproccess_value_counts, left_on='Case ID', right_on='DS Case ID', suffixes=(False, False))
    dfResults = dfResults.merge(activity_value_counts, left_on='Case ID', right_on='Act Case ID', suffixes=(False, False))

    dfResults = dfResults.fillna(0)


#     filterout = len(dfResults[dfResults['last_event_nr']!=cutoff])
#     print("Filterout {} cases because they have less than {} events".format(filterout, cutoff))
    dfResults = dfResults[dfResults['last_event_nr']<=cutoff]
    dfResults.drop(columns=['DS Case ID', 'Act Case ID', 'last_event_nr'], axis=1, inplace=True)
    dfResults = dfResults[dfResults['StartDate'].dt.year == 2016]
   

    return dfResults
```

```python
def make_train_test_split(cutoff, write = False):
    train = setActivityCutoffPointTrain(cutoff)
    test = setActivityCutoffPointTest(cutoff)
    train, test = train.align(test, axis=1, fill_value=0)
    if write:
        train.to_csv("C:/Users/nlsvee/Documents/JADS Y2/PROM/Assignment 2/process-mining-assignment-2/dataset/aggregated/Train {} Events - Case Aggeration.csv".format(cutoff))
        test.to_csv("C:/Users/nlsvee/Documents/JADS Y2/PROM/Assignment 2/process-mining-assignment-2/dataset/aggregated/Test {} Events - Case Aggeration.csv".format(cutoff))
    return train, test
```
